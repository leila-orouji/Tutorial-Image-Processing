```

import matplotlib.pyplot as plt
from skimage import data, color
import matplotlib.image as mpimg
from skimage.color import rgb2gray
```

```
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray

rockt_image = data.rocket();
rocket_grayscaled = rgb2gray(rockt_image)
plt.imshow(rocket_grayscaled)
plt.title('grayscaled')
plt.axis('off')
plt.show()
plt.imshow(rockt_image)
plt.title('RGB')
plt.axis('off')
plt.show()
```

```
# Colors with NumPy
# Obtaining the red values of the image 
red = rockt_image[:, :, 0] 

# Obtaining the green values of the image 
green = rockt_image[:, :, 1] 

# Obtaining the blue values of the image 
blue = rockt_image[:, :, 2]

# Visualizing images
def show_image(image, title='Image', cmap_type='gray'): 
  plt.imshow(image, cmap=cmap_type) 
  plt.title(title) 
  plt.axis('off') 
  plt.show()

# Show original and resulting image to compare
show_image(rockt_image, 'Original', cmap_type='gray')
show_image(red, 'red', cmap_type='gray')
show_image(green, 'green', cmap_type='gray')
show_image(blue, 'blue', cmap_type='gray')
```

```
# Using NumPy for images : Flipping , Extract and analyze features (histogram)
import numpy as np

# Loading the image using Matplotlib 
# madrid_image = plt.imread('/madrid.jpeg') 


# Accessing the shape/size/type of the image 
rockt_image = data.rocket();
rockt_image.shape
rockt_image.size
type(rockt_image)
```

```
# Flip the image in up direction 
vertically_flipped = np.flipud(rockt_image) 
show_image(vertically_flipped, 'Vertically flipped image')

# Flip the image in left direction 
horizontally_flipped = np.fliplr(rockt_image) 
show_image(horizontally_flipped, 'Horizontally flipped image')
```

```
# Applications of histogram: 
# analysis, thresholding, brightness and contrast, equalize an image
plt.hist(rockt_image.ravel(), bins=70)

# Red color of the image 
red = rockt_image[:, :, 0] 
# Obtain the red histogram 
plt.hist(red.ravel(), bins=256)
```
